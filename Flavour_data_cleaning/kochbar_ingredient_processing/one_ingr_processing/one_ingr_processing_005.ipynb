```python
import pandas as pd
import numpy as np
```

```python
done_ingr_001=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_001.csv',index_col=0)
done_ingr_002=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_002.csv',index_col=0)
done_ingr_003=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_003.csv',index_col=0)
done_ingr_004=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_004.csv',index_col=0)
done_ingr_005=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_005.csv',index_col=0)
done_ingr_006=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_done_006.csv',index_col=0)
```

```python
done_ingr_005.drop('tag',axis=1,inplace=True)
done_ingr_006.drop('tag',axis=1,inplace=True)
done_ingr_005.drop('en_ingr',axis=1,inplace=True)
done_ingr_006.drop('en_ingr',axis=1,inplace=True)
```

```python
done_ingr_001['lemma_ingr']=done_ingr_001['parse_ingr']
done_ingr_002['lemma_ingr']=done_ingr_002['parse_ingr']
done_ingr_003['lemma_ingr']=done_ingr_003['parse_ingr']
done_ingr_004['lemma_ingr']=done_ingr_004['parse_ingr']
```

```python
print(done_ingr_001.shape)
print(done_ingr_002.shape)
print(done_ingr_003.shape)
print(done_ingr_004.shape)
print(done_ingr_005.shape)
print(done_ingr_006.shape)
```

```python
one_ingr_done=pd.concat([done_ingr_001,done_ingr_002,done_ingr_003,done_ingr_004,done_ingr_005,done_ingr_006],axis=0,ignore_index=True)
one_ingr_done.shape
```

```python
one_ingr_done=one_ingr_done[one_ingr_done['lemma_ingr']!='']
one_ingr_done['lemma_ingr']=one_ingr_done['lemma_ingr'].str.strip()
one_ingr_done.dropna(inplace=True)
one_ingr_done.reset_index(drop=True,inplace=True)
one_ingr_done.shape
```

```python
recipe_info=pd.read_csv('~/work/food_flavor/kochbar_ingredient_preprocessing/datasets/one_ingr_data/one_ingr_004.csv',index_col=0)
recipe_info.shape
```

```python
recipe_info['recipe_id']=recipe_info['recipe_id'].astype('int')
```

```python
recipe_info
```

```python
recipe_one_done=pd.merge(recipe_info,one_ingr_done,on='ingredient_name',how='inner')
recipe_one_done.shape # why the rows duplicated?
```

```python
recipe_one_done.head(10)
```

```python
recipe_one_done.drop_duplicates(inplace=True)
recipe_one_done.reset_index(drop=True,inplace=True)
recipe_one_done.shape
```

```python
recipe_one_done.sort_values('recipe_id',inplace=True)
recipe_one_done.reset_index(drop=True,inplace=True)
```
